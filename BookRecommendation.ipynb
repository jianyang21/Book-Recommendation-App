```python
from langchain_community.llms import LlamaCpp
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import gradio as gr

# Load the local GGUF model
llm = LlamaCpp(
    model_path="/root/.cache/huggingface/hub/models--TheBloke--Mistral-7B-Instruct-v0.1-GGUF/snapshots/731a9fc8f06f5f5e2db8a0cf9d256197eb6e05d1/mistral-7b-instruct-v0.1.Q4_K_M.gguf",
    temperature=0.7,
    max_tokens=512,
    top_p=0.95,
    repeat_penalty=1.1,
    n_threads=4,
    verbose=False
)

# Prompt Template
prompt = PromptTemplate.from_template("""
You are a book expert. Based on the user's interest: "{user_interest}", recommend 5 great books and include a 1-line reason for each.
""")

# Chain
chain = LLMChain(llm=llm, prompt=prompt)

# Gradio function
def recommend_books(user_input):
    return chain.run(user_interest=user_input)

# Gradio UI
interface = gr.Interface(
    fn=recommend_books,
    inputs=gr.Textbox(lines=2, placeholder="e.g., I love thrillers and fantasy"),
    outputs=gr.Textbox(label="üìö Book Recommendations"),
    title="üìñ Offline Book Recommender (Mistral 7B GGUF)",
    description="Uses locally quantized Mistral model. No internet or tokens needed."
)

interface.launch()
```
